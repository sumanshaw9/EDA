1. What is the average rating for each product category?

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce') #converting rating column to numeric for further calculation
df.info()
```

```python
df_clean = df.dropna(subset=['category', 'rating'])
df_clean.head(1)
```

```python
avg_rating_by_cat = df_clean.groupby('category')['rating'].mean().sort_values(ascending=False)
avg_rating_by_cat
```

2. What are the top rating_count products by category?

```python
top_rating_count_by_category = df_clean.groupby('category').apply(lambda x: x.sort_values('rating_count', ascending=False)).head(1)
top_rating_count_by_category
```

3. What is the distribution of discounted prices vs. actual prices?

```python
df_clean.info()
```

```python
# Step 1:Remove non-numeric characters (e.g., ₹, commas, etc.)
df_clean['actual_price'] = df_clean['actual_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df_clean['discounted_price'] = df_clean['discounted_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)

# Step 2:Convert to float
df_clean['actual_price'] = pd.to_numeric(df_clean['actual_price'], errors='coerce')
df_clean['discounted_price'] = pd.to_numeric(df_clean['discounted_price'], errors='coerce')

# Step 3:Preview
print(df_clean[['actual_price', 'discounted_price']].head())
plt.figure(figsize=(10,5))
sns.kdeplot(df_clean['actual_price'].dropna(), alpha=0.6, label='Actual Price')
sns.kdeplot(df_clean['discounted_price'].dropna(), alpha=0.6, label='Discounted Price')
plt.legend()
plt.title('Distribution: Discounted Price vs Actual Price')
plt.xlabel('Price')
plt.xlim(0, 10000)
plt.ylabel('Frequency')
plt.show()
```

4. How does the average discount percentage vary across categories?

```python
df_clean['discount_percentage'] = df_clean['discount_percentage'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df_clean['discount_percentage']
```

```python
df_clean['discount_percentage'] = pd.to_numeric(df_clean['discount_percentage'], errors='coerce')

avg_discount = df_clean.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)
print(avg_discount)
```

```python
#Top 15 categories with highest average discount
top_discounted = avg_discount.head(15)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_discounted.values, y=top_discounted.index, palette='viridis')
plt.title("Top 15 Categories by Average Discount")
plt.xlabel("Average Discount Percentage")
plt.ylabel("Category")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.show()
```
