1. What is the average rating for each product category?

```python
avg_rating_by_cat = df.groupby('category')['rating'].mean()
```

2. What are the top rating_count products by category?

```python
top_rating_count_by_category = df.groupby('category').apply(lambda x: x.sort_values('rating_count', ascending=False)).head(1)
top_rating_count_by_category
```

3. What is the distribution of discounted prices vs. actual prices?

```python
df.columns
```

```python
sns.barplot(data = df, x= 'discounted_price', y='actual_price')
plt.xticks(rotation=45)
plt.title("Discounted Price Vs Actual Price")
plt.show()
```
