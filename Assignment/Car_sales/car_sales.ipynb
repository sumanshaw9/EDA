1. What is the average selling price of cars for each dealer, and how does it compare across different dealers ?

```python
df.head(1)
```

```python
avg_price_by_dealer = df.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending= False)
avg_price_by_dealer
```

```python
#Comparision using barplot
avg_price_by_dealer = avg_price_by_dealer.reset_index()
avg_price_by_dealer
```

```python
#visualization for comparision between dealer and selling price
plt.figure(figsize= (10,8))
sns.barplot(data = avg_price_by_dealer, x = 'Dealer_Name', y = 'Price ($)', palette='viridis')
plt.title('Average Selling Price by Dealer Type')
plt.ylabel('Average Selling Price (INR)')
plt.xlabel('Seller Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
trends?

```python
#using standard deviation as a measure of variation
price_variation = df.groupby('Company')['Price ($)'].std().sort_values(ascending=False)

print(price_variation.head(10))  # Top 10 brands with highest variation
```

```python
top_brands = price_variation.head(5).index  # Top 5 variable brands

plt.figure(figsize=(10, 8))
sns.boxplot(data=df[df['Company'].isin(top_brands)], x='Company', y='Price ($)')
plt.title('Price Variation Among Top 5 Most Variable Car Brands')
plt.ylabel('Price ($)')
plt.xlabel('Company')
plt.yscale('log')  # If prices vary widely, log scale helps
plt.tight_layout()
plt.show()
```

3.What is the distribution of car prices for each transmission type, and how do the interquartile ranges 
compare?

```python
# Plot Distribution of Selling Prices by Transmission Type  This boxplot will visually show:
#Median prices
#Outliers
#IQR (box height)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Transmission', y='Price ($)')
plt.title('Distribution of Car Prices by Transmission Type')
plt.xlabel('Transmission')
plt.ylabel('Selling Price (INR)')
plt.yscale('log')  # Use log scale if prices vary widely
plt.tight_layout()
plt.show()
```

```python
#calculating Interquartile ranges
def calc_iqr(series):
    return series.quantile(0.75) - series.quantile(0.25)

iqr_values = df.groupby('Transmission')['Price ($)'].apply(calc_iqr)
print("IQR of Selling Price by Transmission Type:\n")
print(iqr_values)
```

6. How does the average selling price of cars vary by customer gender and annual income?

```python
# Bin the income into categories
df['Income Bracket'] = pd.cut(df['Annual Income'], bins=[0, 30000, 60000, 90000, 120000], labels=['<30K', '30K-60K', '60K-90K', '90K-120K'])
# Use seaborn for cleaner visualization
sns.barplot(data=df, x='Income Bracket', y='Price ($)', hue='Gender', ci=None)

plt.title("Avg Car Price by Gender and Income Bracket")
plt.ylabel("Average Price ($)")
plt.xlabel("Income Bracket")
plt.show()
```

12. Is there any seasonal trend in car sales based on the date of sale?

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.month_name()
sns.countplot(data=df, x='Month', order=pd.date_range('2022-01-01', periods=12, freq='M').strftime('%B'))
plt.title("Seasonal Trend in Car Sales")
plt.xticks(rotation=45)
plt.show()
```
